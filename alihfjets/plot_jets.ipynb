```python
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
jets = joblib.load('run_fj.py_output.joblib')
```

```python
jets.describe()
```

```python
sns.pairplot(data=jets)
```

```python
f, axes = plt.subplots(1,2, figsize=(10,5), sharex=True, sharey=True)
cmap = sns.cubehelix_palette(start=0, light=1, as_cmap=True)
sns.kdeplot(jets['pt'], jets['eta'], cmap=cmap, shade=True, ax=axes[0])
sns.kdeplot(jets['ptsub'], jets['eta'], cmap=cmap, shade=True, ax=axes[1])
```
