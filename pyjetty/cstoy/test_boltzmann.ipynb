```python
import ROOT
```

```python
import cstoy
```

```python
be = cstoy.BoltzmannEvent(multiplicity=100, max_pt=10)
```

```python
for i in range(1000):
    be.generate()
```

```python
tc = ROOT.TCanvas("test_boltzmann","test_boltzmann", 600, 600)
tc.Divide(2,2)
tc.cd(1)
ROOT.gPad.SetLogy(1)
ROOT.gPad.SetLogx(1)
be.histogram_pt.Sumw2()
be.histogram_pt.Scale(1./be.nEvent/be.histogram_pt.GetBinWidth(1))
be.histogram_pt.Draw()
tc.cd(2)
be.histogram_eta.Scale(1./be.nEvent/be.histogram_eta.GetBinWidth(1))
be.histogram_eta.Draw()
tc.cd(3)
be.histogram_phi.Scale(1./be.nEvent/be.histogram_phi.GetBinWidth(1))
be.histogram_phi.Draw()
tc.SetLogy(1)
tc.Draw()
```
